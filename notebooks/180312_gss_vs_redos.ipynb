```python
import pandas as pd
from collections import Counter
import os
import os.path as op
from viruscope.orf_setup import prodigal


wd = "/mnt/scgc/simon/simonsproject/gss_vs/"
```

```python
df = pd.read_csv("/mnt/scgc/simon/simonsproject/info/GORG_SAG_summary_180305.txt", sep="\t")
```

```python
Counter(["-".join(i.split("-")[:2]) for i in df['SAG']])
```

```python
to_fix = df[(df['RS_group_notes'] == 'Chisholm old assemblies') & (df['Database'] == 'GSS')]['SAG']
```

```python
for sag in to_fix:
    sagcode = op.join(wd, "*",'{}*'.format(sag))
    !rm {sagcode}
```

```python
err_prots = []
out_proteins = []

for name in list(to_fix):
    c = op.join("/mnt/scgc/simon/simonsproject/gss_contigs/contigs/{}_contigs.fasta".format(name))
    output = "/mnt/scgc/simon/simonsproject/gss_vs/prodigal/"
    outfiles = [os.path.join(output, "prodigal", name + "_proteins.fasta"),
                os.path.join(output, "prodigal", name + "_genes.fasta"),
                os.path.join(output, "prodigal", name + ".gbk"),
                os.path.join(output, "prodigal", name + ".scores")]
    try:
        p_proteins, p_genes, p_genbank, p_score = prodigal(c, outfiles, verbose=True)
    except Exception as inst:
        print(inst)
        err_prots.append(name)
        continue
    out_proteins.append(p_proteins)    
```
